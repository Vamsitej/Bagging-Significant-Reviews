```python
gionee_like_reviews_total=[]
gionee_reviews=[]
mac =[]

from bs4 import BeautifulSoup

installation_synonyms_list=[]  
url=("https://www.synonym.com/synonyms/delivery")
response = requests.get(url)
soup = bs(response.content,"html.parser")# creating soup object to iterate over the extracted content 
#reviews = soup.find(class_="css-1kg1yv8 eh475bn0")    # Extracting the content under specific tags  
reviews = soup.find(class_="results-container")
for review in reviews:
  installation_synonyms_list.append(review)
```
